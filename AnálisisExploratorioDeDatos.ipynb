**Plot**

* Para este tipo de gráficos se utiliza Seaborn, que permite generar un histograma para variables categóricas. Cada barra en el gráfico de conteo representa una clase.
* Se crea un bucle para el conteo y distribución de las clases. La sentencia select_dtypes(include = ’object’) selecciona las columnas categóricas con sus valores y las muestra.
* Se incluye también If para elegir solo las tres columnas con 10 o menos clases usando series.**nunique() < 10**.


```python
for col in DatosMelbourne.select_dtypes(include='object'):
  if DatosMelbourne[col].nunique()<10:sns.countplot(y=col, data=DatosMelbourne)
  plt.show()
```

**5) Agrupación por variables categóricas**





```python
for col in DatosMelbourne.select_dtypes(include='object'):
  if DatosMelbourne[col].nunique() < 10:
    display(DatosMelbourne.groupby(col).agg(['mean']))
```

### **Paso 4: Identificación de relaciones entre pares variables**

Una matriz de correlaciones es útil para analizar la relación entre las variables numéricas.
Se emplea la función corr()

```python
DatosMelbourne.corr()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(DatosMelbourne.corr(), cmap='RdBu_r', annot=True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
MatrizInf = np.triu(DatosMelbourne.corr())
sns.heatmap(DatosMelbourne.corr(), cmap='RdBu_r', annot=True, mask=MatrizInf)
plt.show()
```

* Una correlación es un valor entre -1 y 1 que equivale a qué tan cerca se mueven simultáneamente los valores de dos variables. 
* Una correlación positiva significa que a medida que una característica aumenta, la otra también aumenta.
* Una correlación negativa significa que a medida que una característica disminuye, la otra también disminuye. 
* Las correlaciones cercanas a 0 indican una relación débil, mientras que las más cercanas a -1 o 1 significan una relación fuerte.


## **Preparación de los datos**

 ### *Limpiando valores nulos*

```python
DatosMelbourne=DatosMelbourne.dropna()
DatosMelbourne
```

### *Limpiando datos atípicos*

#### *Después de limpiar los valores nulos, verificamos si siguen existiendo los valores atípicos que habíamos identificado.*

```python
sns.boxplot('YearBuilt', data=DatosMelbourne)
plt.show()
```

#### *Al confirmar que sigue el valor más incoherente encontrado en la columna de 'YearBuilt', procedemos a recortar dicho valor.*

```python
DatosMelbourne = DatosMelbourne[DatosMelbourne['YearBuilt'] > 1200]
```

#### *Observamos con los diagramas de caja los valores atípicos de las otras variables y corroboramos si quitamos el valor atípico de 'YearBuilt'*

```python
VariablesValoresAtipicos = ['Price', 'Landsize', 'BuildingArea', 'YearBuilt']
for col in VariablesValoresAtipicos:
  sns.boxplot(col, data=DatosMelbourne)
  plt.show()
```

#### *Confirmamos la eliminación del valor atípico de 'YearBuilt' y quitamos los valores atípicos de 'BuildingArea' y 'Landsize'.*

```python
DatosMelbourne = DatosMelbourne[DatosMelbourne['BuildingArea'] < 2000]
DatosMelbourne = DatosMelbourne[DatosMelbourne['Landsize'] < 30000]
```

```python
VariablesValoresAtipicos = ['Price', 'Landsize', 'BuildingArea']
for col in VariablesValoresAtipicos:
  sns.boxplot(col, data=DatosMelbourne)
  plt.show()
```

#### *Debido a que "aparentemente" las distribuciones de las tres variables que nos faltaban por limpiar son muy parecidas, no se decidió recortar ningún valor en la variable 'Price'.*

### *Limpiando relaciones entre pares de variables*

#### *Como se había apreciado en el mapa de calor, la variable 'Bedroom2' estaba fuertemente correlacionada con la variable 'Rooms', por lo que se decidió quitar la variable 'Bedroom2'*

```python
DatosLimpios = DatosMelbourne.drop(columns='Bedroom2')
```

#### *Confirmamos en el mapa de calor*

```python
plt.figure(figsize=(14,7))
MatrizInf = np.triu(DatosLimpios.corr())
sns.heatmap(DatosLimpios.corr(), cmap='RdBu_r', annot=True, mask=MatrizInf)
plt.show()
```

*Como podemos observar, después de limpiar todos nuestros datos de los valores nulos y los datos atípicos, ahora tenemos nuevas correlaciones en nuestro conjunto. No obstante, a pesar de que las variables 'BuildingArea' presenta una correlación de aproximadamente 0.66 con la variable 'Rooms' y algunas otras candidatas a presentar una correlación fuerte como 'Price' o 'Bathroom', ninguna de éstas cumple con el criterio de los tercios superiores e inferiores, por lo que no se eliminará ninguna otra columna*
